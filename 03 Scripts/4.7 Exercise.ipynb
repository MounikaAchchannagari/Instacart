```python
# Creating User defined price_label function
```

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Applying price_label function to each row in df Dataframe
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# value counts in price_range coloumn
```

```python
# Creating for loop for checkecking each row in orders_day_of_week column
```

```python
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
ords_prods_merge['busiest_day'] = result
```

```python
# Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days.
```

```python
result = []

for value in ords_prods_merge["orders_day_of_week"]:
    if value in [0, 1]:
        result.append("Busiest days")
    elif value in [3, 4]:
        result.append("Slowest days")
    else:
        result.append("Regularly busy")
```

```python
# Assigning result to ords_prods_merge['busiest day'] column
```

```python
ords_prods_merge['busiest_days'] = result
```

```python
ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

```python
# creating for loop for checking each row in order_hour_of_day column
```

```python
result = []

for value in ords_prods_merge["order_hour_of_day"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        result.append("Most orders")
    elif value in [17, 8, 18, 19, 20, 7, 21, 22]:
        result.append("Average orders")
    else:
        result.append("Fewest orders")
```

```python
# Assigning result to ords_prods_merge['busiest_period_of_day'] column
```

```python
ords_prods_merge['busiest_period_of_day'] = result
```
