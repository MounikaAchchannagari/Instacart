```python
# Question 6:  If the mean of the prices of products purchased by a user is lower than 10, then flag them as a â€œLow spender.â€
# If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a â€œHigh spender.â€
```

```python
# Create mean product price using transform() function for each user.
```

```python
ords_prods_merge['mean_product_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Derivig columns using loc function
# creating spender flag usin loc
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# If the median of â€œdays_since_prior_orderâ€ is higher than 20, then the customer should be labeled a â€œNon-frequent customer.â€
# If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a â€œRegular customer.â€
# If the median is lower than or equal to 10, then the customer should be labeled a â€œFrequent customer.â€
```

```python
# Use groupby and transform to find median number of days between orders for each user
```

```python
ords_prods_merge['median_days_between_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(5)
```

```python
# Derivig columns using loc function
# creating order frequency flag usin loc
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] > 20, 'order_frequency'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_between_orders'] <= 20) & (ords_prods_merge['median_days_between_orders'] > 10), 'order_frequency'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
ords_prods_merge['order_frequency'].value_counts(dropna = False)
```
